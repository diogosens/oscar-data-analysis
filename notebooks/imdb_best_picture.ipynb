```python
# importing libraries

import pandas as pd
import urllib.request, urllib.parse, urllib.error
import json
```

```python
imdb_best_picture_list = pd.read_csv("Best Picture Oscar Nominees by Year (Academy Award.csv")
```

```python
nominees = {
    'imdb_id' : imdb_best_picture_list["Const"],
    'movie' : imdb_best_picture_list["Title"],
    'oscar_winner' : imdb_best_picture_list["Description"].notnull().astype('int')
    }
nominees_dataframe = pd.DataFrame(nominees)
nominees_dataframe
```

```python
with open('APIkeys.json') as f:
    keys = json.load(f)
    omdbapi = keys['OMDBapi']
    
service_url = 'http://www.omdbapi.com/?i='
apikey = '&apikey=' + omdbapi
```

```python
def search_movie(imdb_id):
    url = service_url + imdb_id + apikey
    uh = urllib.request.urlopen(url)
    data = uh.read()
    json_data=json.loads(data)
    return json_data
```

```python
movie_info = {}
for index in range(len(nominees_dataframe)):
    movie = nominees_dataframe['movie'].iloc[index]
    imdb_id = nominees_dataframe['imdb_id'].iloc[index]
    movie_info[imdb_id] = search_movie(imdb_id)
    print(f'Retrieving the data of {movie} now...')
```

```python
omdb_dataframe = pd.DataFrame(movie_info).T.reset_index()
```

```python
oscar_best_movie_nominees = pd.concat([omdb_dataframe, nominees_dataframe['oscar_winner']], axis=1)
oscar_best_movie_nominees
```

```python
rename_columns = {
    'index' : 'IMDbID',
    'oscar_winner' : 'OscarWinner'
}

oscar_best_movie_nominees.rename(columns = rename_columns, inplace = True)
```

```python
oscar_best_movie_nominees.to_csv("oscar_best_movie_nominees.csv", index_label = False)
```
