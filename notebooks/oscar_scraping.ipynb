```python
# importing libraries

import bs4
import urllib.request as urllib_request
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.request, urllib.parse, urllib.error
import json
```

```python
# reading AMPAS page dataset

url = "oscar_page.html" # manually get from https://awardsdatabase.oscars.org/
page = open(url, encoding = "utf8")
soup = BeautifulSoup(page.read())
```

```python
# finding all the movie nominess by edition and category

oscar_nominations = soup.find_all('a', {'class' : "nominations-link"})

oscar = {}
edition = ''
category = ''
movie = ''

for item in oscar_nominations:
    item_content = item.get_text()
    
    if item_content[-1] == ')' and item_content[:4].isnumeric():
        edition = item_content
        oscar[edition] = {}
    
    if 'categoryExactId' in item.get('href'):
        category = item_content
        oscar[edition][category] = {}
        movie_list = []
        nominee_list = []
        
    if 'filmId' in item.get('href'):
        movie = item_content
        movie_list.append(movie)
        oscar[edition][category] = movie_list
```

```python
# creating DataFrame

oscar_dataframe = pd.DataFrame(oscar).T
oscar_dataframe.to_csv('oscar_dataframe.csv', index_label = False)
```
