```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
```

```python
top_box_office_imdb_id = pd.read_csv("top_box_office_imdb_id.csv")
```

```python
box_office_list = []
year_list = []

for imdb_id in top_box_office_imdb_id['IMDbID']:
    url = f"https://www.boxofficemojo.com/title/{imdb_id}"
    response = urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    
    original_ballot = (soup.find('div', {'class' : 'a-section a-spacing-none a-spacing-top-base'})
                        .find('tr', {'class' : 'mojo-revenue-source-theatrical'}))
    
    if original_ballot != None:
        box_office_list.append(int(original_ballot.find('span', {'class' : 'money'})
                       .get_text().replace('$', '').replace(',', '')))
        
    else:
        box_office_list.append(int(soup.find('span', {'class' : 'money'})
                       .get_text().replace('$', '').replace(',', '')))
    
    year_list.append(soup.find('span', {'class' : 'a-size-large a-color-secondary'})
                     .get_text().replace(' ', '')[1:-1])
    
top_box_office_imdb_id['BoxOffice'] = box_office_list
top_box_office_imdb_id['Year'] = year_list
```

```python
top_box_office_imdb_id.to_csv("top_box_office.csv", index_label = False)
```
